```python
import whisper

print("Downloading and loading the Whisper 'base' model...")
# This will download the model weights to your Colab instance (~140 MB)
asr_model = whisper.load_model("base")
print("Model loaded successfully!")
```

```python
from google.colab import files
import os

print("Please upload your 1-minute audio file:")
# 1. Open an upload widget in Colab
uploaded = files.upload()

if uploaded:
    # Get the name of the uploaded file
    audio_file_name = list(uploaded.keys())[0]
    print(f"\nProcessing '{audio_file_name}'...")

    # 2. Run the transcription
    # Setting fp16=False to avoid CPU warnings and use FP32
    result = asr_model.transcribe(audio_file_name, fp16=False)

    # 3. Output the final text
    print("\n" + "="*50)
    print("📝 TRANSCRIPTION OUTPUT:")
    print("="*50)
    print(result["text"].strip())
    print("="*50)
```
