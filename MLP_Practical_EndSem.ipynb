```python
import numpy as np
from itertools import permutations

def relu(z):
    """Simple ReLU activation function."""
    return np.maximum(0, z)

def mlp_forward(x, W, b, v, c):
    """Computes the forward pass of a single hidden layer MLP."""
    hidden = relu(np.dot(W, x) + b)
    return np.dot(v, hidden) + c
```

```python
n_input = 4
d_hidden = 3
n_output = 1

# Set seed for reproducibility
np.random.seed(42)

# Randomly initialize the network parameters (representing a theoretical local minimum)
W = np.random.randn(d_hidden, n_input)
b = np.random.randn(d_hidden)
v = np.random.randn(n_output, d_hidden)
c = np.random.randn(n_output)

# Random input vector
x = np.random.randn(n_input)

# Calculate the baseline output
y_original = mlp_forward(x, W, b, v, c)
print(f"Baseline Output: {y_original[0]:.8f}")
```

```python
# Generate permutation matrices from an identity matrix
identity_matrix = np.eye(d_hidden)
all_permutations = list(permutations(identity_matrix))

print(f"Testing all {len(all_permutations)} parameter permutations:\n")
print("-" * 50)

# Apply symmetry transformations
for i, p_tuple in enumerate(all_permutations):
    P = np.array(p_tuple)

    # Transform parameters according to the mathematical proof
    W_prime = np.dot(P, W)
    b_prime = np.dot(P, b)
    v_prime = np.dot(v, P.T)
    # The output bias 'c' remains untouched

    # Calculate the new output with the permuted weights
    y_prime = mlp_forward(x, W_prime, b_prime, v_prime, c)

    # Calculate the absolute difference to prove equivalence
    diff = np.abs(y_original - y_prime)[0]

    print(f"Permutation {i+1}: Output = {y_prime[0]:.8f} | Difference = {diff:.2e}")
```
